```python
Task 2: Gender Distribution
```

```python
import matplotlib.pyplot as plt

# Extract gender information (Assume 'Gender' is the column name)
gender_distribution = data['gender'].value_counts()

# Plot a pie chart for gender distribution
gender_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')  # Remove the label
plt.show()

# Alternatively, you can plot a bar chart:
gender_distribution.plot(kind='bar')
plt.title('Gender Distribution')
plt.ylabel('Count')
plt.show()
```

```python
Task 3: Descriptive Statistics for Numerical Columns
```

```python
# Identify numerical columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Calculate mean, median, and standard deviation for each numerical column
for col in numerical_columns:
    mean = data[col].mean()
    median = data[col].median()
    std_dev = data[col].std()
    
    print(f"Column: {col}")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Standard Deviation: {std_dev}")
    print('-' * 50)
```

```python
Task 8: Investment Duration
```

```python
# Define mapping from text to numbers (in years)
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

# Apply the mapping
data['Investment Duration (Numeric)'] = data['Duration'].map(duration_mapping)
# Now compute the average and median
mean_duration = data['Investment Duration (Numeric)'].mean()
median_duration = data['Investment Duration (Numeric)'].median()

print(f"Average Investment Duration: {mean_duration:.2f} years")
print(f"Median Investment Duration: {median_duration:.2f} years")
import seaborn as sns

sns.histplot(data['Investment Duration (Numeric)'], bins=5, kde=True, color='teal')
plt.title('Investment Duration (Converted to Numeric)')
plt.xlabel('Investment Duration (Years)')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()

```

```python
Task 10: Correlation Analysis
```

```python
# Duration mapping (convert to years)
duration_map = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
data['Duration_Numeric'] = data['Duration'].map(duration_map)

# Expectation mapping (example: convert text to scores)
expect_map = {
    'High returns': 5,
    'Moderate returns': 3,
    'Safety of capital': 2,
    'Regular income': 4,
    'Tax saving': 3
}
data['Expect_Numeric'] = data['Expect'].map(expect_map)
# Select only relevant numeric columns
corr_cols = [
    'age', 'Mutual_Funds', 'Equity_Market', 'Debentures',
    'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold',
    'Duration_Numeric', 'Expect_Numeric'
]

# Create correlation matrix
correlation_matrix = data[corr_cols].corr()

# Display correlation matrix
print(correlation_matrix)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Investment Variables')
plt.tight_layout()
plt.show()
```
